```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, Dense, GlobalMaxPooling2D
from tensorflow.keras.applications import VGG16
```

```python

def build_model(): 
    input_layer = Input(shape=(100,100,3))
    
    vgg = VGG16(include_top=False)(input_layer)

    # Classification Model  
    f1 = GlobalMaxPooling2D()(vgg)
    class1 = Dense(10000, activation='relu')(f1)
    class2 = Dense(1, activation='sigmoid')(class1)
    
    # Bounding box model
    f2 = GlobalMaxPooling2D()(vgg)
    regress1 = Dense(10000, activation='relu')(f2)
    regress2 = Dense(12, activation='sigmoid')(regress1)
    
    facetracker = Model(inputs=input_layer, outputs=[class2, regress2])
    return facetracker
```

```python
facetracker = build_model()
```

```python
facetracker.summary()
```

```python
batches_per_epoch = int(2000)/10
lr_decay = (1./0.75 -1)/batches_per_epoch
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.0001, decay=lr_decay)
```

```python
classloss = tf.keras.losses.BinaryCrossentropy()
regressloss = tf.keras.losses.CategoricalCrossentropy()
```

```python
facetracker.compile(opt,classloss,regressloss)
```

```python
from mtcnn import MTCNN

# Create an instance of the MTCNN class
mtcnn = MTCNN()

# Load an image from file
cap = cv2.VideoCapture(-1)
while cap.isOpened():
    success, image = cap.read()
    if not success:
        print("Ignoring empty camera frame.")
        continue
        

    # Detect faces and facial landmarks in the image
    results = mtcnn.detect_faces(image)
    # The results variable will be a list of dictionaries, each dictionary representing one face
    # in the image. The dictionary will contain the bounding box coordinates of the face, as well as
    # the facial landmarks, represented as x, y coordinates
    for face in results:
        bounding_box = face['box']
        facial_landmarks = face['keypoints']

        x, y, width, height = bounding_box
        cv2.rectangle(image, (x, y), (x+width, y+height), (255, 0, 0), 2)

        for key, value in facial_landmarks.items():
            cv2.circle(image, value, 2, (0, 255, 0), 2)
        if cv2.waitKey(5) & 0xFF == 27:
            cv2.destroyAllWindows()
            break
    if cv2.waitKey(5) & 0xFF == 27:
        cv2.destroyAllWindows()
        cap.release()
        break

cv2.imshow("Facial Landmarks", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from mtcnn import MTCNN

# Create an instance of the MTCNN class
mtcnn = MTCNN()

# Open a connection to the camera
cap = cv2.VideoCapture(0)

while True:
    # Get a frame from the camera
    ret, frame = cap.read()

    # Detect faces and facial landmarks in the frame
    results = mtcnn.detect_faces(frame)

    # Draw rectangles and circles on the frame to mark the bounding box and landmarks
    for face in results:
        bounding_box = face['box']
        facial_landmarks = face['keypoints']

        x, y, width, height = bounding_box
        cv2.rectangle(frame, (x, y), (x+width, y+height), (255, 0, 0), 2)

        for key, value in facial_landmarks.items():
            cv2.circle(frame, value, 2, (0, 255, 0), 2)

    # Show the frame
    cv2.imshow("Facial Landmarks", frame)

    # Exit the loop if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close the window
cap.release()
cv2.destroyAllWindows()
```
